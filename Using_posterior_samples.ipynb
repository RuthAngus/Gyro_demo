# A demonstration of how to use posterior samples to make an age prediction.

```python
import numpy as np
import matplotlib.pyplot as plt
import triangle
import h5py
%matplotlib inline
```

Load the samples

```python
with h5py.File("thinned_samples.h5", "r") as f:
    samples = f["samples"][:, :, :] 
    nwalkers, n, ndim = samples.shape
    flatchain = samples.reshape((-1, ndim))  # flatchain has shape (nsamples, ndim)
    mcmc_result = map(lambda v: (v[1], v[2]-v[1], v[1]-v[0]),
                      zip(*np.percentile(flatchain, [16, 50, 84], axis=0)))
    result = np.array(mcmc_result)[:, 0]
print "median parameter values = ", result
print "Gyrochronology parameters (a, n, b) = ", result[:3]
```

Make a triangle plot

```python
fig = triangle.corner(flatchain[:, :3], labels=["$a$", "$n$", "$b$"])
```

The model (calculates an age, given a rotation period and colour). 
pars is a 2d array containing the posterior samples of a, n & b (in that order), p is rotation period and bv is B-V.

```python
def age_model(pars, p, bv):
    return 10**((np.log10(p) - np.log10(pars[:, 0]) - pars[:, 2]*np.log10(bv - .45)) / pars[:, 1]) / 1000.
```

This function calculates the age with posterior samples. It returns the median age, the 16th and 84th percentile uncertainties and the age samples.

```python
def age_calc(bv, bv_err, p, p_err, flatchain):
    nsamp = len(flatchain)
    bv_samp = bv_err * np.random.randn(nsamps) + bv  # draw samples from Gaussian with mu = bv and sigma = bv_err.
    p_samp = p_err * np.random.randn(nsamps) + p     # draw samples from Gaussian with mu = p and sigma = p_err.
    age_samples = age_model(flatchain[:, :3], p_samp, bv_samp)
    p = np.percentile(age_samples, [16, 50, 84])
    return np.array([p[1], p[2]-p[1], p[1]-p[0]]), age_samples
```

Take the Sun as an example:

```python
p, p_err = 26.9, .1
bv, bv_err = 0.65, .01
age, age_samples = age_calc(bv, bv_err, p, p_err, flatchain)
print "age = ", age[0], "+", age[1], "-", age[2]

plt.hist(age_samples, 50, color="w")
plt.axvline(age[0], color="r", label="Median age = %.2f Gyrs" % age[0])
plt.axvline(age[0]+age[1], color="k", ls="--")
plt.axvline(age[0]-age[2], color="k", ls="--")
plt.xlabel("Age (Gyr)")
plt.legend()
```

If you don't use posterior samples and assume everything is Gaussian you get a very different answer, which doesn't account for the covariances between parameters!

```python
def age_no_samples(pars, bv, bv_err, p, p_err, nsamps=10000):
    bv_samp = bv_err * np.random.randn(nsamps) + bv  # draw samples from Gaussian with mu = bv and sigma = bv_err.
    p_samp = p_err * np.random.randn(nsamps) + p     # draw samples from Gaussian with mu = p and sigma = p_err.
    age_samples = 10**((np.log10(p_samp) - np.log10(pars[0]) - pars[2]*np.log10(bv_samp - .45)) / pars[1]) / 1000.
    p = np.percentile(age_samples, [16, 50, 84])
    return np.array([p[1], p[2]-p[1], p[1]-p[0]]), age_samples
```

```python
pars = result[:3]
age, age_samples = age_no_samples(pars, bv, bv_err, p, p_err)
print "age = ", age[0], "+", age[1], "-", age[2]

plt.hist(age_samples, 50, color="w")
plt.axvline(age[0], color="r", label="Median age = %.2f Gyrs" % age[0])
plt.axvline(age[0]+age[1], color="k", ls="--")
plt.axvline(age[0]-age[2], color="k", ls="--")
plt.xlabel("Age (Gyr)")
plt.legend()
```
